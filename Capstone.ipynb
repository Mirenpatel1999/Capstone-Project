```python
import pandas as pd
import numpy as np
```

```python
DS = pd.read_csv("data_sources.csv")
JA = pd.read_csv("job_attachments.csv")
J = pd.read_csv("jobs.csv")
SCA = pd.read_csv("Status Codes Attachments.csv")
SC = pd.read_csv("Status Codes.csv")
UA = pd.read_csv("Users Attachments.csv")
U = pd.read_csv("Users.csv")
ST = pd.read_csv("Scrap Types.csv")
SE = pd.read_csv("scrap_events.csv")
status_code = {1:"UPTIME",2:"DOWNTIME"}
RD = pd.read_csv("raw_machine_data.csv")
assets = {2842:"Press 2",2962:"Press 1"}
```

```python
SC_final = SC[2:]
SC_final['Machine Status'] = SC_final['parent_id'].map(status_code)
SC_final.rename(columns = {"id":"status_code_id"}, inplace=True)
SC_final.head()

```

```python
SCA_final = pd.merge(SCA,SC_final,on="status_code_id")
SCA_final.drop(columns=['id','status_code_id',"parent_id",'color'], inplace=True)
SCA_final.rename(columns = {"name":"reason"}, inplace=True)
```

```python
SCA_final.to_excel("Status Code Final.xlsx",index=False)
```

```python
J.rename(columns = {"id":"job_id"}, inplace=True)
JA_final = pd.merge(JA,J,on="job_id")
JA_final.drop(columns=['id','job_id'], inplace=True)
JA_final.to_excel("Jobs Final.xlsx",index=False)
```

```python
U.rename(columns = {"id":"user_id"}, inplace=True)
UA_final = pd.merge(UA,U,on="user_id")
UA_final.drop(columns=['id','user_id'], inplace=True)
UA_final.to_excel("Users Final.xlsx",index=False)
```

```python
ST.rename(columns = {"id":"scrap_type_id"}, inplace=True)
SE_final = pd.merge(SE,ST,on="scrap_type_id")
SE_final.drop(columns=['id','scrap_type_id'], inplace=True)
SE_final.to_excel("Scrap Final.xlsx",index=False)
```
